```python
df.drop(columns=['College_ID'], inplace=True)
```

```python
x = df.drop(columns=['Placement'])
y = df['Placement']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.3, random_state=0)
```

```python
df['Internship_Experience'].unique()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
```

```python
tnf1 = ColumnTransformer([
    ('Internship_Experience', OrdinalEncoder(categories='auto'), [4])
],remainder='passthrough')
```

```python
encodefeture = tnf1.fit_transform(x_train)
```

```python
tnf3 = LogisticRegression()
```

```python
from sklearn.pipeline import make_pipeline
```

```python
pipe = make_pipeline(tnf1, tnf3)
```

```python
pipe.fit(x_train, y_train)
```

```python
y_pred = pipe.predict(x_test)
```

```python
accuracy_score(y_test, y_pred)
```
