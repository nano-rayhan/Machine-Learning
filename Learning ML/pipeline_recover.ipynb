```python
import numpy as np
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer 
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, LabelEncoder, StandardScaler, MinMaxScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn.feature_selection import chi2, SelectKBest
```

```python
df = pd.read_csv('college_student_placement_dataset.csv')
```

```python
df.head(5)
```

```python
df.columns
```

```python
df.drop(columns=['College_ID'],inplace=True)
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.drop(columns=['Placement']),
                                                    df['Placement'],
                                                    test_size=0.3,
                                                    random_state=0)
y_train.shape
```

```python
tnf1 = ColumnTransformer(transformers=[
    ('orde', OrdinalEncoder(categories=[['No','Yes']]), [4]),
    ('ohe', OneHotEncoder(sparse_output=False, handle_unknown='ignore'), [0,1,2,3,5,6,7])
],remainder='passthrough')
```

```python
tnf2 = DecisionTreeClassifier()
```

```python
pipe = make_pipeline(tnf1,tnf2)
```

```python
pipe.fit(x_train, y_train)
```

```python
y_pred = pipe.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
