```python
import numpy as np
import pandas as pd
```

```python
time = pd.read_csv('messages.csv')
date = pd.read_csv('orders.csv')
```

```python
date.info()
```

Date handling

```python
date['date'] = pd.to_datetime(date['date'])
```

```python
date['date_year']  = date['date'].dt.year
```

```python
date.sample(5)
```

```python
date['date_month'] = date['date'].dt.month
date['date_day'] = date['date'].dt.day
```

```python
date.sample(5)
```

```python
date['date_month_name'] = date['date'].dt.month_name()
date.sample(5)
```

```python

date['date_week_name'] = date['date'].dt.day_name()
date['date_is_weekend'] = np.where(date['date_week_name'].isin(['Saturday','Friday']),1,0)
date.head(6)
```

```python
date.drop(columns=['product_id','city_id','orders'],inplace=True)
```

```python
date['quarter'] = date['date'].dt.quarter
date.sample(6)
```
